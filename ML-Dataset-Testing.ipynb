# Preprocessing The Dataset

# 

# Drop Unwanted Attributes

```python
data=read.drop(['frame.time','icmp.unused','http.file_data','http.request.uri.query','http.referer','http.request.full_uri',
                 'http.request.version','http.tls_port','dns.qry.name.len','dns.qry.type','dns.retransmit_request_in',
                 'mqtt.conack.flags','mqtt.msg_decoded_as','mqtt.msg','mbtcp.len','mbtcp.trans_id','mbtcp.unit_id'],axis=1)
data.info()
```

# Drop Null/ Missing Values

```python
data.dropna(inplace=True)
```

# Checking for Null Values

```python
data.isnull().sum()
```

# Correlation Among The Attributes

```python
data.corr(method ='pearson')
```

```python
data.corr().style.background_gradient(cmap='YlOrBr')
```

# Finding Unique Classes

```python
data["Attack_type"].unique()
```

# Count Records per Class

```python
data["Attack_type"].value_counts()
```

# Label Encoding the Dataset

```python
categorical=[features for features in data.columns if data[features].dtype=="object"]
categorical

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()


for features in categorical:
    #label_encoder=LabelEncoder()
    data[features]=label_encoder.fit_transform(data[features])
```

```python
data.info()
```

# Count Records per Class

```python
data["Attack_type"].value_counts()
```

# Count Columns where Data Type != Object

```python
numerical_features=[feature for feature in data.columns if data[feature].dtype!='object']
len(numerical_features)
```

# Dependent/ Independent Attributes

```python
Y=data[['Attack_type']]
```

```python
X=data.drop(['Attack_type'],axis='columns')
```

# Balancing Dataset using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
x,y = oversample.fit_resample(X,Y)
```

```python
y.value_counts()
```

# Splitting Dataset into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=15,train_size=0.70)
```

```python
print('Shape of X_train set {}'.format(x_train.shape),'&',' Size of Y train set {}'.format(y_train.shape))
```

```python
print('Shape of X_test set {}'.format(x_test.shape),'&',' Size of Y_test set {}'.format(y_test.shape))
```

# NaÃ¯ve Bayes

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
model=MultinomialNB()
```

# Naive Bayes Training Set

```python
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
train_predictions=model.predict(x_train)
test_predictions=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print("Naive Bayes Training Report")
print(classification_report(y_train,train_predictions))
```

# Naive Bayes Testing Set

```python
print("Naive Bayes Test Report")
print(classification_report(y_test,test_predictions))
```

# Confusion Matrix of Naive Bayes Classifier

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, test_predictions)
print(cf_matrix)
```

```python
import seaborn as sns
sns.heatmap(cf_matrix, annot=True)
```

# Precision Naive Bayes Classifier

```python
from sklearn.metrics import precision_score
test_precision=precision_score(y_test, test_predictions,average='weighted')
test_precision
```

# Recall Naive Bayes Classifier

```python
from sklearn.metrics import recall_score
test_precision2=recall_score(y_test, test_predictions,average='weighted')
test_precision2
```

# F1-Score Naive Bayes Classifier

```python
from sklearn.metrics import f1_score
test_precision3=f1_score(y_test, test_predictions,average='weighted')
test_precision3
```
